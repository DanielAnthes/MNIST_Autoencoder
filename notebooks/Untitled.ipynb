# Autoencoder

```python
import tensorflow as tf
import sys
sys.path.append('../src/')
from adv_autoencoder import AAE
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import mnist
import random
import util
```

```python
weights = '../models/mnist_aae20210224-223235'

model = AAE()
model.load(weights)
```

```python
(X_train, label_train), _ = mnist.load_data()
X_train = tf.cast(X_train, tf.float32) / 255
X_train = X_train[:,:,:,None]
X_train.shape
```

## Example Reconstructions

```python
X = X_train[:5,:,:,:]
Z, R = model(X)
reconstruction = R.numpy()

fig = util.plot_reconst(X, reconstruction)
plt.show()
```

## Latent Space

```python
Z, _ = model(X_train)
Z = Z.numpy()
fig = util.plot_latent_space(Z, label_train)
plt.show()
```
